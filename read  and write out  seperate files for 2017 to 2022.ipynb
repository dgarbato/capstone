### function to get data for certain years from Iowa Liquor Sales file

```python
import csv
def getstuff2(filename, years):
    dict1={}
    with open(filename, "r") as csvfile:
        datareader = csv.reader(csvfile)
        dict1['header'] = next(datareader)  # the header row
        count = 0
        rows = []
        for row in datareader:
#             if row[2][-4:] in years:
            if row[1][-4:] in years:
                rows.append(row)
                count += 1
                if count % 1000 == 0:
                    print(row[1][-4:])
           
        dict1['rows'] = rows
        
        print(count)
                # done when having read a consecutive series of rows 
        return dict1
```

### function to write out data

```python

def write(filename,dict1):
    with open(filename, 'w', newline="") as file:
        csvwriter = csv.writer(file) # 2. create a csvwriter object
        csvwriter.writerow(dict1['header']) # 4. write the header
        csvwriter.writerows(dict1['rows']) # 5. write the rest of the data
```
