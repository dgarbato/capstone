```python
def replacer(series):
    if series.dtype == 'int64':
        return series
    else:
        series = series.str.replace(',','')
        series = series.str.replace('-','0')
        return series
    
```

```python
# census[['Total_population','Male','Female']].apply(lambda x: x.str.replace(',','')).astype(float)
census = census.apply(lambda x: replacer(x))
```

```python
def dashit(series):
    if series.dtype == 'int64':
        return False
    return np.any(series.str.find('-') != -1)

    
```

```python
census.columns[census.apply(lambda x: dashit(x))]
```

```python
def chktype(series):
    try: 
        series = series.astype(float)
        return series
    except:
        return series
```

```python
census = census.apply(lambda x: chktype(x))
```
