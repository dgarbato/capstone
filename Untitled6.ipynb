```python
clust_dict ={'Cheap':'Moderately Priced', 'High_End':'High End', 'Moderately_High':'Moderately High', 
             'Most_Variety':'Most Variety', 'Popular_Bottles':'Popular Bottles','Quick_Shots':'Quick Shots',
            'Volume_For_Dollar':'Volume For Dollar'}


store_trans_cluster['New_Cluster_Name'] = store_trans_cluster['cluster_name'].map(clust_dict)
```

```python
store_trans_cluster.groupby(['cluster_name'])['Bottle_Retail'].median().round(2)
```

```python
store_trans_cluster.groupby(['cluster_name']).agg({'Item_Number':['nunique'],'revenue':['sum'],'Bottle_Volume':['min','max','median'],'Bottle_Retail':['min','max', 'median'], 'Bottles_Sold':['sum', 'median'],'Num_Stores_Selling':['sum', 'median']})
```

```python
feature = 'Bottle_Retail'

def box(feature,ylabel):

    sns.set(rc = {'figure.figsize':(15,8)})
    box = sns.boxplot(x='New_Cluster_Name', y=feature, data=store_trans_cluster,palette='Blues')

    box.set_xlabel('Product Cluster',fontsize=15)
    box.set_ylabel(ylabel,fontsize=15)
```

```python
box('Bottle_Retail','State Bottle Retail Price')
```
