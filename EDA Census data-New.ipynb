# sort by population per store

Total dollars this year (year 1): (# of people >21yo) * (Average/median income in Zip) = (Total Dollar Size of Market)
mkt share = 1 /(# of existing stores + new_store)

```python

# [1/(stores+1) if stores !=0 else None for stores in census_combined4.number_of_stores]
census_combined4['market_share'] = [1/(stores+1) if stores !=0 else None for stores in census_combined4.number_of_stores]
```

```python
def new_store(pop21,med_income,pct_drinkers=0.70,pct_disposable_alcohol=0.01,pct_disposable_inc=0.50,
              mkt_share=0.25,avg_profit_margin=0.15):
    total_addressable_income = pop21 * med_income * pct_drinkers
      
    total_addressable_disp_inc = total_addressable_income * pct_disposable_inc
    
    mkt_size= total_addressable_disp_inc*pct_disposable_alcohol  #number_dollars_alcolhol spent in mkt
    
    revenue_new_store = mkt_size *  mkt_share
    
    profit_new_store = revenue_new_store * avg_profit_margin
    
    return round(profit_new_store)
```

```python
census_combined4['profit_to_store'] = new_store(census_combined4['21_years_and_over'],census_combined4['Median_household_income_(dollars)'],
         mkt_share=census_combined4['market_share'])
```
