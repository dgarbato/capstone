##### create function for EDA of chain stores

```python
def summary_func2(casey):
    cols_list = ['store_revenue',
     'tot_num_unique_items',
     'num_unique_clusters',
     
     'rev_Cheap',
     'rev_High_End',
     'rev_Moderately_High',
     'rev_Most_Variety',
     'rev_Popular_Bottles',
     'rev_Quick_Shots',
     'rev_Volume_For_Dollar',
 

                 
     'cl_Cheap',
     'cl_High_End',
     'cl_Moderately_High',
     'cl_Most_Variety',
     'cl_Popular_Bottles',
     'cl_Quick_Shots',
     'cl_Volume_For_Dollar']
    features =[]
    median_values=[]
    mean_values = []
    std_dev = []


    for col in cols_list:
        features.append(col)

        median = casey[col].median()
        median_values.append(median)

        mean = casey[col].mean()
        mean_values.append(mean)
        
        std_ = casey[col].std()
        std_dev.append(std_)

    summary_df = pd.DataFrame({'features':features, 'median_values':median_values, 'mean_values':mean_values, 'std_deviation':std_dev})
    summary_df['median_values'] = summary_df['median_values'].round(2)
    summary_df['mean_values'] = summary_df['mean_values'].round(2)
    summary_df['std_deviation'] = summary_df['std_deviation'].round(2)
    return summary_df
```
