```python
new_categories=[]
for categ in categories['Category_Name']:
    if 'whiskey liqueur' in categ.lower():
        new_categories.append('Whiskey Liqueur')
    elif 'tequila' in categ.lower():
         new_categories.append('Tequila')
    elif 'cocktail' in categ.lower():
        new_categories.append('Cocktails')
    elif 'american distilled' in categ.lower():
        new_categories.append('American Distilled Spirits Specialty')
    elif 'imported distilled' in categ.lower():
        new_categories.append('Imported Distilled Spirits Specialty')
    elif 'temporary' in categ.lower():
        new_categories.append('Temporary & Specialty Packages')
    elif 'bourbon' in categ.lower():                               #  in ['bottled in bond bourbon','single barrel bourbon whiskies','straight bourbon whiskies']:
        new_categories.append('Bourbon')
    elif 'scotch' in categ.lower() :                       #in ['scotch whiskies', 'single malt scotch']:
        new_categories.append('Scotch')
    elif 'brandies' in categ.lower() :                     #in ['imported brandies', 'american brandies']:
        new_categories.append('Brandies')
    elif 'vodka' in categ.lower():
        new_categories.append('Vodkas')
    elif 'rum' in categ.lower() :
         new_categories.append('Rum')
    elif 'liqueur' in categ.lower() :
         new_categories.append('Liquers')
    elif 'gin' in categ.lower() :
         new_categories.append('Gin')
    elif 'schnapps' in categ.lower():                                                # categ.lower().find('schnapps')!= -1:
         new_categories.append('Schnapps')
    elif  'whiskies' in categ.lower():
        new_categories.append('Whiskies')
    
        
    else:
        new_categories.append(categ)

new_categories

nc_set= set(new_categories)
sorted(list(nc_set))        
        
    
        
        
    
```
